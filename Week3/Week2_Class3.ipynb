Now we will go a layer deeper to understand what is exactly scikit-learn Linear Regressor doing under the hood.

So scikit-learn is obviously reducing some kind of loss. And it being at a higher loss it comes to a lower loss and once it hits the minima of the loss it stops.

The loss it optimizes is called OLS(Ordinary least squares) loss.

Consider linear model = $$\hat{Y}_i = B_o + B_1 * X_i (Predicted value)$$. As the linear estimate of $$Y_i (target value)$$

Then loss at any point i is given by $$(Y_i - \hat{Y}_i)^2$$ or $$(Y_i - (B_o + B_1 * X_i))^2$$
At all points , the loss is just the summation

#### This is the vertical distance between actual Y value at i and predicted Y value at i

The value which we are trying to alter in order to reduce the loss is **B_o** and **B_1** , we cant change **X_i** as it is fixed

So lets plot loss L as a function of $$B_o , B_1$$ , $$L(B_o , B_1)$$


```python
B_o, B_1 = np.random.randn(1000) , np.random.randn(1000)
X = np.random.randn(100)
Y = 0.3 + 0.5*X
z = [sum((Y  - a - b * X)**2)/100 for a,b in zip(B_o,B_1)]
trace1 = go.Scatter3d(
    x=B_o,
    y=B_1,
    z=z,
    mode='markers',
    marker=dict(
        size=12,
        line=dict(
            color='rgba(217, 217, 217, 0.14)',
            width=0.5
        ),
        opacity=1
    )
)

data = [trace1]
layout = go.Layout(
    title = "The Graph",
    scene= go.Scene(
        xaxis=go.XAxis(title='B_o'),
        yaxis=go.YAxis(title='B_1'),
        zaxis=go.ZAxis(title='Loss')
    )
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

So sklearn and its **optimizer**, tries to reach the very bottom of the above graph and find corresponding values of B_o and B_1. But how do we do it? Suppose we start at some random values B_o & B_1 (might be the exactly the ones we want, but probably is almost zero). How do we reach the optimum value? Cant just keep guessing, right?


So there is this algorithm called Gradient Descent algorithm. So this simply descents downwards the above loss function. In which direction? Down obviously. But note what is the gradient( slope) of function w.r.t B_o and B_1 at the minima? It is zero for both. So it is obvious that we do some damn thing to get to zero gradient. So what we do it follow the slope and go in its direction. And keep updating B_o and B_1 in the process. Find the slope at that point and go in its direction. It works, we will reach the bottom doing this in theory( Practice has various disturbances and some things we need to watch out). But how much do we move down the function given the slope direction?. It depends on how fast we want to reach at the minima and many other factors. We will cover that in next classes. But keep this in mind.

```python
#So we do this in code and update B_o and B_1 by reducing the value in negative slope direction.
#Why negative? Quite intuitive.
#Note this is function in two variables therefore we use partial derivatives for each coefficient.
B_o = B_o - ((-2/n)*sum(Y - B_o - B_1 * X)) #Can get from wolfram alpha
B_1 = B_1 - ((-2/n)*sum((Y - B_o - B_1 * X) * X)) #Can get from wolfram alpha
```

So it is nothing but a for loop that we will implement in the direction of descent. Until values do not change.
Note, once the bottom is reached updates wont happend as the gradient(second half of rhs) will become zero.
That is called convergence.
But until convergence (some_times might not converge) you repeat the updates. Or may be if you are fairly near convergence you stop.

But gradient descent can be fooled if there are more than one minimas (local minimas) , then the minima reached will depend on where B_1 is initialized. Like in the below if B_1 is **?** then it will reach global minima but if it is **?**
then it will reach local minima. Therefore random initizalization is very necessary and must be repeated at times

```python
B_1 = np.random.randn(1000)
X = np.random.randn(1000)
y = [sum((100 + (b**2)*X - (b * X))**2)/100 for b in B_1] # Y = 100 + (b^2)*X
trace1 = go.Scatter(
    x=B_1,
    y=y,
    mode='markers',
    marker=dict(
        size=12,
        line=dict(
            color='rgba(217, 217, 217, 0.14)',
            width=0.5
        ),
        opacity=1
    )
)

data = [trace1]
layout = go.Layout(
    title = "The Graph",
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```
