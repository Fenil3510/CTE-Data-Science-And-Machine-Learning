```python
#Note it is also inside the sklearn package you have in your local machine
from sklearn.datasets import load_boston
load_boston = load_boston()
```

```python
type(load_boston) #Sklearn can also handle, data , but Pandas is widely used and better.
```

```python
df = pd.DataFrame(np.concatenate([load_boston.data, np.expand_dims(load_boston.target,axis =1)] , axis = 1) , columns= list(load_boston.feature_names) + ['MV'] ) #Converted to pandas
```

```python
df.head()
```

Data wrangling is shaping or modelling the data the way you want it. Some example below

```python
#Multiply each row in column CRIM by 2
df['CRIM'] = df['CRIM']*2 #Scalar multiplication
```

```python
df.head()
```

```python
df['Add_two'] = df['CRIM'] + df['ZN']
```

```python
df.head(20)
```

```python
#Filtering data according to requirements make a copy or else
df2 = df.copy()
```

```python
df3 = df #Watch what happens
```

```python
df3['CRIM'] = 0
```

```python
df3.head()
```

```python
df.head() #Changed the main df along with itself
```

```python
#But
df2['MV'] = df2['MV'] * 5
```

```python
df2.head()
```

```python
df.head() #not affected
```

```python
#Now back to filtering
df2 = df2[df2['MV'] > 50]
```

```python
df2.head()
```

```python
df2 = df2[(df2['MV'] > 100) & (df2['LSTAT'] < 20)]
```

```python
df2.head()
```

```python
id(df2)
```

```python
#Changing values based on condition
df2.loc[df2['ZN'] == 0 , "CRIM"] = 13
```
