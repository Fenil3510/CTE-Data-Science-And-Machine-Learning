Correlation plots can be useful. Correlation is the ratio of change in one variable when other is changed(Almost similar to slope)
Formula :(Sample Correlation)


Let x, y be two variables with n data points each.

Pearson(widely used) Correlation = $$r = \frac{\sum_{i=1}^{n}(x_i - \bar{x})(y_i - \bar{y})}{\sqrt[]{\sum_{i=1}^{n}(x_i - \bar{x})^2 \sum_{i=1}^{n}(y_i - \bar{y})^2}}$$


```python
correlate = df.corr()
```

```python
correlate.shape
```

```python
mask = np.zeros_like(correlate, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
```

```python
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(5, 5))
```

```python
sns.heatmap(correlate,vmax=.3, ax= ax, mask = mask, center= 0, linewidths=.5)
display(f)
```

```python
#3d Plot
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(xs = df['CRIM'],ys = df['AGE'] , zs = df['MV'],zdir = 'z')
ax.set_xlabel('CRIM')
ax.set_ylabel('AGE')
ax.set_zlabel('MV')
display(fig)
```

```python
#Interactive 3d plot using Plotly
#Please do not waste time mugging up the interface and commands , you get used to it via practice.
x, y, z = df['CRIM'] , df['AGE'] , df['MV']
trace1 = go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=12,
        line=dict(
            color='rgba(217, 217, 217, 0.14)',
            width=0.5
        ),
        opacity=0.8
    )
)

data = [trace1]
layout = go.Layout(
    title = "The Graph",
    scene= go.Scene(
        xaxis=go.XAxis(title='CRIM'),
        yaxis=go.YAxis(title='AGE'),
        zaxis=go.ZAxis(title='MV')
    )
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

Ok now lets use sklearn library to quickly fit our linear model

$$MV_i = B_o + B_1 * CRIM_i + B_2 * ZN_i + B_3 * INDUS_i + ......... + B_{13} * LSTAT_i$$

So we will be regressing CRIM,ZN,INDUS,....,LSTAT on MV.
Ie. Determining coefficients B_o, B_1...etc. by some method.
Method is minimizing the loss between


$$Y_i$$ and $$\hat{Y}_i$$


For each $$i$$ from 0 to len(dataset)

ie. $$\frac{\sum_{i = 0}^{n}((Y_i - \hat{Y}_i)^2)}{n} $$

```python
from sklearn.linear_model import LinearRegression
```

```python
??LinearRegression
```

```python
lm_model = LinearRegression(fit_intercept= True, n_jobs = 1, normalize = False)
```

```python
lm_model.fit(df.drop(['MV'], axis = 1), df['MV'])
```

```python
lm_model.intercept_
```

```python
lm_model.coef_
```

```python
coef_dataframe = pd.DataFrame(list(zip(df.columns.values, lm_model.coef_)))
```

```python
coef_dataframe.columns = ['Variable', 'Coefficient']
```
